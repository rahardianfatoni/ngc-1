## **IMPORT LIBRARIES AND DATA**

```python
import pandas as pd
```

```python
raw_data = pd.read_csv('https://raw.githubusercontent.com/FTDS-learning-materials/phase-0/main/src/data_ngc.csv')
raw_data
```

## **DATA CLEANING AND FILTERING**

### Finding Dtypes and Non-Null Count for each column

```python
raw_data.info()
```

### Removing empty columns



```python
raw_data.drop(['fips', 'admin2'], axis=1, inplace=True)
```

### Removing empty rows

```python
raw_data[raw_data['active'].notna()]
```

```python
raw_data.dropna(subset=['province_state'], inplace=True)
```

### Drop 'Unknown' values in province_state

```python
no_unknown_data = raw_data[raw_data['province_state'] != 'Unknown']
```

## **DATA EXPLORATION**

### Group data by province_state and show by sum of confirmed, deaths, and recovered

```python
province_data = no_unknown_data.groupby('combined_key')['confirmed', 'deaths', 'recovered'].sum()
province_data
```

```python
country_data = raw_data.groupby('country_region')['confirmed', 'deaths', 'recovered'].sum()
country_data
```
