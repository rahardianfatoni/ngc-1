```python
from google.colab import auth
from google.cloud import bigquery
import pandas as pd

auth.authenticate_user()
project_id = "proud-woods-408708" # Ini project ID
client = bigquery.Client(project=project_id)
```

```python
query = """SELECT * FROM patents-public-data.uspto_oce_cancer.publications"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job.head()
```

```python
query = """SELECT *
FROM patents-public-data.uspto_oce_cancer.INFORMATION_SCHEMA.COLUMNS"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```

```python
query = """SELECT PARSE_DATE('%Y%m%d', Grant_or_Publication_Date) AS Grant_or_Publication_Datetime, Patent_Title
FROM `patents-public-data.uspto_oce_cancer.publications`
WHERE EXTRACT(YEAR FROM PARSE_DATE('%Y%m%d', Grant_or_Publication_Date)) BETWEEN 2012 AND 2023
ORDER BY EXTRACT(YEAR FROM PARSE_DATE('%Y%m%d', Grant_or_Publication_Date)) DESC
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job.head()
```

```python
query = """SELECT LEFT(Grant_or_Publication_Date, 4) AS Year, SUBSTRING(Grant_or_Publication_Date, 5, 2) AS Month, COUNT(1) AS Total_Patents
FROM `patents-public-data.uspto_oce_cancer.publications`
WHERE CAST(LEFT(Grant_or_Publication_Date, 4) AS INTEGER) BETWEEN 2012 AND 2023
AND LOWER(Patent_Title) LIKE "%data mining%"
GROUP BY LEFT(Grant_or_Publication_Date, 4), SUBSTRING(Grant_or_Publication_Date, 5, 2)
"""
# Set up the query
query_job = client.query(query).to_dataframe()
query_job
```
